```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('iris.csv')
print(df.head())
```

```python
print(df.describe())
```

```python
# You can clean 'Sepal_Length'
df.drop(['Sepal_Length'] , inplace=True , axis = 1)
```

```python
df.groupby('Class')[' Petal_Length'].mean()
```

```python

ax = sns.countplot(data = df , x = ' Sepal_Width', hue = 'Class', palette='Set2')
ax.set(xlabel = 'Width of the sepal' , ylabel = 'Number of Flowers' , title = 'Width of Species')
```

```python
interval=[0,1,2,4]
categories = ['<1', '1-2', '>2']
df['category'] = pd.cut(df[' Petal_Width'],interval, labels=categories)
ax = sns.countplot(data=df,x='category',hue='Class')
ax.set(xlabel='Classes', ylabel='Count')
```

```python

ax = sns.countplot(df[df['Class']!='Iris-setosa'] , x = ' Sepal_Width', hue = 'Class' , palette = 'Set3')
ax.set(xlabel = 'Variety' , ylabel = 'Count')
```
